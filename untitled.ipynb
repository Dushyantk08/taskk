```python
import pandas as pd
df1=pd.read_csv('/content/purchase_behaviour.csv')
df2=pd.read_csv('/content/transaction_data.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
new_df=pd.merge(df1,df2,how='inner',on='LYLTY_CARD_NBR')
```

```python
new_df.head()
```

```python
top_3=new_df.groupby('PROD_NAME')['TOT_SALES'].sum().nlargest(3)
top_3
```

```python

loyal_customers = new_df.groupby("LYLTY_CARD_NBR")["TOT_SALES"].sum().nlargest(10)

loyal_profile = new_df[new_df["LYLTY_CARD_NBR"].isin(loyal_customers.index)][["LIFESTAGE", "PREMIUM_CUSTOMER"]].value_counts()

print("\nCharacteristics of Loyal Customers:\n", loyal_profile)
```
